```python
#Caracteristicas generales, seguridad social en salud y educacion
#Links a los conjuntos de datos
url_cgsse = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/cgsse.CSV'
url_ocupados ='https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Ocupados.CSV'
url_ft = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Fuerza%20de%20trabajo.CSV'
url_desocupados = 'https://raw.githubusercontent.com/niconomist98/DataAnalyticsUQ/refs/heads/main/Datos/GEIH2025/No%20ocupados.CSV'

# CGSSE
df_cgsse = pd.read_csv(url_cgsse, encoding='latin-1', sep=';', 
                       low_memory=False)

#Ocupados
df_ocupados=pd.read_csv(url_ocupados, encoding='latin-1',sep=';',
                        low_memory=False)

#Fuerza de trabajo
df_ft=pd.read_csv(url_ft,
                  encoding='latin-1',sep=';',low_memory=False)
#No ocupados
df_desocupados=pd.read_csv(url_desocupados,encoding='latin-1',
                           sep=';',low_memory=False)
```

```python
codigos = ['P3271','DPTO','P6040', 'P6050', 'P3041','P2057','P6170', 'P3042','P3039']

nombres_columnas = {
    'P3271': 'sexo',
    'DPTO': 'departamento',
    'P6040': 'edad',
    'P6050': 'parentesco',
    'P3041': 'institucion',
    'P2057': 'nivel_educativo',
    'P6170': 'estado_civil',
    'P3042': 'tipo_institucion',
    'P3039': 'como_se_reconoce',
}

# Si quieres usar solo las columnas que están en df_cgsse
codigos_encontrados = [c for c in codigos if c in df_cgsse.columns]

df_seleccion = df_cgsse[codigos_encontrados].rename(columns=nombres_columnas)
df_seleccion.head(100)
```

```python
df_seleccion["nivel_educativo"].unique()
```

```python
df_seleccion["institucion"].unique()
```

```python
# Estadísticas descriptivas por columna
estadisticas = pd.DataFrame({
    'media': df_seleccion.mean(numeric_only=True),
    'mediana': df_seleccion.median(numeric_only=True),
    'desviacion_estandar': df_seleccion.std(numeric_only=True),
    'max': df_seleccion.max(numeric_only=True),
    'min': df_seleccion.min(numeric_only=True),
    'nulos': df_seleccion.isna().sum(),
})

modas = df_seleccion.apply(lambda x: x.mode().iloc[0] if len(x.mode()) > 0 else np.nan)
estadisticas['moda'] = modas

estadisticas
```

```python
# Filtrar los registros con edad menor a 25
N_df = df_seleccion[df_seleccion['edad'] < 25]

N_df.head()
```

```python
# Estadísticas descriptivas por columna
estadisticas = pd.DataFrame({
    'media': N_df.mean(numeric_only=True),
    'mediana': N_df.median(numeric_only=True),
    'desviacion_estandar': N_df.std(numeric_only=True),
    'max': N_df.max(numeric_only=True),
    'min': N_df.min(numeric_only=True),
    'nulos': N_df.isna().sum(),
})

modas = N_df.apply(lambda x: x.mode().iloc[0] if len(x.mode()) > 0 else np.nan)
estadisticas['moda'] = modas

estadisticas
```
